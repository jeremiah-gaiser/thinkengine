```python
import json 
import numpy as np
from glob import glob
import re
import math
import matplotlib.pyplot as plt
```

```python
exp1_data_dir = '../exp_data/exp1_3/'
exp1_data_files = glob(exp1_data_dir + "*.json")
exp1_data = [-1 for _ in range(99)]

for data_f in exp1_data_files:
    with open(data_f) as df:
        trial_idx = int(re.search(r'_(\d+).json', data_f).groups()[0])-1
        exp1_data[trial_idx] =  json.load(df)
        
exp1_data = exp1_data[:exp1_data.index(-1)] 
```

```python
def get_odds_seq(trial_data, background_prob):
    odds_seq = []
    
    for pos_score, neg_score in zip(trial_data['pos_scores'], trial_data['neg_scores']):
        total_score = pos_score + neg_score

        if total_score == 0:
            continue
        
        if pos_score == 0: 
            odds_seq.append(0)
            continue

        if neg_score == 0:
            correct_prob = 1
        else:
            correct_prob = pos_score/total_score
        
        odds_seq.append(correct_prob/background_prob)

    return odds_seq

def running_average(seq, w):
    convolved_seq = []
    for i in range(len(seq)-w):
        convolved_seq.append(sum(seq[i:i+w]) / w)
    return convolved_seq
    

hyper_param_vals = []
trial_seqs = []
mean_convergence = []

for trial in exp1_data:
    trial_seqs.append(get_odds_seq(trial, 1/9))
    hyper_param_vals.append([trial['threshold'], trial['reward'], trial['penalty'], trial['explore_variance'], trial['refractory_step']])
    mean_convergence.append(sum(trial_seqs[-1][-10:])/len(trial_seqs[-1][-10:])) 

plt.plot((running_average(mean_convergence, 5)))
```

```python
cmap = plt.cm.viridis
running_avg_seqs = [running_average(t, 5) for t in trial_seqs]

# Create a line plot
plt.figure(figsize=(10, 6))
for i, trial in enumerate(running_avg_seqs):
    color = cmap(float(i) / len(running_avg_seqs))
    plt.plot(trial, label=f'Trial {i+1}', color=color)
    
plt.title('Trial Data Over Time')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
